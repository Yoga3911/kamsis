```python
# GENERATE KEY

import sympy
import math

p1 = sympy.randprime(2 ** 4, 2 ** 8)
p2 = sympy.randprime(2 ** 4, 2 ** 8)
n = p1 * p2

print("N =", n)

pn = (p1 - 1) * (p2 - 1)
e = sympy.randprime(1, pn)
while math.gcd(e, pn) > 1:
  e = sympy.randprime(1, pn)
print("Public Key = ", e)

d = (pn + 1) / e
i = 1
while d % 1 > 0:
  i += 1
  d = (i * pn + 1) / e
d = int(d)
print("Private Key =", d)
```

```python
# RSA

import base64
import math

print("Enter N: ")
n = input()
n = int(n)
l = math.ceil(math.log2(n))
print("Enter Public Key: ")
e = input()
e = int(e)
print("Enter Plain Text: ")
plain = input()

plain_byte = bytearray(plain, "utf-8")
chiper_binary = ""
for x in plain_byte:
  x = x ** e % n
  chiper_binary += format(x, "0" + str(l) + "b")
while len(chiper_binary) % 8 > 0:
  chiper_binary += "0"

chiper_byte = bytearray()
for x in range(0, len(chiper_binary), 8):
  tmp = chiper_binary[x:x + 8]
  tmp = int(tmp, 2)
  chiper_byte.append(tmp)
chiper_base64 = base64.b64encode(chiper_byte)
print("Encrypted Message: ", chiper_base64.decode("ascii"))
```

```python
# RSA

import math
import base64

print("Enter N: ")
n = input()
n = int(n)
l = math.ceil(math.log2(n))
print("Enter Private Key: ")
d = input()
d = int(d)
print("Enter Chiper Text: ")
chiper = input()

chiper = chiper.encode("ascii")
chiper_byte = base64.b64decode(chiper)
chiper_binary = ""
for x in chiper_byte:
  chiper_binary += format(x, "08b")

plain = ""
for x in range(0, len(chiper_binary), l):
  x = chiper_binary[x:x + l]
  x = int(x, 2)
  plain += chr(x ** d % n)
print("Plain Text: ", plain)
```

```python
# STEGANO

import math
from PIL import Image

print("Enter the secret message: ")
secret = input()
secret_byte = bytearray(secret, "utf-8")
secret_binary = []
for x in secret_byte:
  secret_binary.append(format(x, "08b"))
secret_binary = "".join(secret_binary)
print("Secret Message: ", secret_binary)

print("Enter the cover image path: ")
cover_path = input()
cover = Image.open(cover_path)
cursor = 0
for y in range(cover.size[1]):
  for x in range(cover.size[0]):
    pixel = cover.getpixel((x, y))
    if cursor < len(secret_binary):
      stego = 2 * math.floor(pixel[2] / 2) + int(secret_binary[cursor])
      cursor = cursor + 1
    else:
      stego = 2 * math.floor(pixel[2] / 2) + 0
    cover.putpixel((x, y), (pixel[0], pixel[1], stego))

print("Enter the stego image path: ")
stego_path = input()
cover.save(stego_path)
```
